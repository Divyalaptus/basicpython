```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("D:/Acadview training notes/titanic train.csv")
df1 = pd.read_csv("D:/Acadview training notes/titanic test.csv")
two_set = [df,df1]

df2 = pd.concat(two_set)
df2.head()
```

```python
df2 = df2.dropna(subset = ["Survived"])
df2.isnull().sum()
```

```python
hasCabin = df2["Cabin"].notnull() #series to store boolean values for the has_cabin
 
df2 = df2.assign(Has_cabin = hasCabin.values)
df2
#df2.append(columns=["Has_cabin"])
```

```python
Titles = df2["Name"].str.split(',',expand = True)
Titles
Titles2 = Titles[1].str.split('.',expand = True)
Titles2
df2 = df2.assign(Title = Titles2[0])
df2
```

```python
lambda x : re.search("([A-Z])[a-z]+)\.",x).group(1)
```

```python
import re
df2["Title"]=df2.Name.apply(lambda x : re.search("([A-Z][a-z]+)\.",x).group(1))
df2.Title.unique()
```

```python
df2['Title'].replace(['Don','Rev', 'Dr', 'Mme', 'Ms',
       'Major', 'Lady', 'Sir', 'Mlle', 'Col', 'Capt', 'Countess',
       'Jonkheer'],'Special')
```

```python
df2 = df2.fillna(df2.median())
df2.isnull().sum()
df2['Cabin'] = df2.fillna(df2['Cabin'].mode()[0])
df2['Embarked'] = df2.fillna(df2['Embarked'].mode()[0])
df2
df2.isnull().sum()
df2.head()
```

```python
categorical_age = pd.qcut(df2["Age"],q=4,labels = None)
categorical_fare = pd.qcut(df2["Fare"],q = 4,labels = None)
df2 = df2.assign(Categorical_age=categorical_age.values)
df2 = df2.assign(Categorical_fare=categorical_fare.values)
df2.head()
```

```python
#Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch
df2 = df2.drop( ['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis=1)
df2.head()
```

```python
df_with_dummies = pd.get_dummies( df2 )
df_with_dummies.head()
```

```python
X = df_with_dummies.drop(["Survived"],axis = 1)
y = df_with_dummies["Survived"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state = 0)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train,y_train)
```

```python

pred = reg.predict(X_test)
result = pd.DataFrame({"Predicted":pred,"Actual":y_test})
result
```
